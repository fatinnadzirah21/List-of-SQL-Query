```python
import pandas as pd
import requests
import json
```

```python
response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=01534b2b788774ab2c62af604febf074&page=3")
data = response.text
parse_json = json.loads(data)
print(data)
```

**json -> dataframe**

```python
df = pd.DataFrame(parse_json)
df
```

```python
df[['adult', 'backdrop_path', 'genre_ids', 'id', 'original_language', 'original_title', 'overiview', 'popularity', 'poster_path', 'release_date', 'title', 'video', 'vote_avg', 'vote_count']] = df["results"].apply(lambda x: pd.Series(str(x).split(", '")))
df.head()
```
